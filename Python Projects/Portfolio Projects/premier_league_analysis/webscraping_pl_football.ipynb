```python
#Get data from fbref.com
import requests
import pandas as pd
from bs4 import BeautifulSoup

pl_standings_url = 'https://fbref.com/en/comps/9/Premier_League_Stats'
data = requests.get(pl_standings_url)
```

```python
#Parsing HTML
soup = BeautifulSoup(data.text)

#Select fist element that has table.stats_table class (standing table)
standings_table = soup.select('table.stats_table')[0]

#find <a> (links) in standings table
links = standings_table.find_all('a')
links = [l.get('href') for l in links]
links = [l for l in links if '/squads/' in l]

#turn links into full urls using format string
team_urls = [f"https://fbref.com{l}" for l in links]
team_urls
```

Extract Match stats using Pandas and Request

```python
first_team_url = team_urls[0]
first_team_data = requests.get(first_team_url)

#get Scores & Fixtures from first team url using pandas
first_team_matches = pd.read_html(first_team_data.text, match= "Scores & Fixtures ")
```

```python
first_team_matches[0].head()
```

```python
soup = BeautifulSoup(first_team_data.text)
links = soup.find_all('a')
links = [l.get('href') for l in links]
links = [l for l in links if l and 'all_comps/shooting/' in l]
links
```

```python
shooting_data = requests.get(f"https://fbref.com{links[0]}")
shooting_data_pd = pd.read_html(shooting_data.text, match="Shooting")[0]
```

```python
shooting_data_pd.head()
```

Cleaning and Merging Scraped Data 

```python
# Drop first index level
shooting_data_pd.columns = shooting_data_pd.columns.droplevel()
shooting_data_pd.head()
```

```python
#merge 2 tables together
team_data = first_team_matches[0].merge(shooting_data_pd[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
team_data.head()
```

Scraping Data For multiple Seasons and Teams with a Loop

```python
years = list(range(2023, 2020, -1))
all_matches = []
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
import time
for year in years:
    data = requests.get(standings_url)
    soup = BeautifulSoup(data.text)
    standings_table = soup.select('table.stats_table')[0]

    links = [l.get("href") for l in standings_table.find_all('a')]
    links = [l for l in links if '/squads/' in l]
    team_urls = [f"https://fbref.com{l}" for l in links]

    #urls for previous_season
    previous_season = soup.select("a.prev")[0].get("href")
    standings_url = f"https://fbref.com{previous_season}"

    for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("-Stats","").replace("-", " ")

        data = requests.get(team_url)
        matches = pd.read_html(data.text, match="Scores & Fixtures")[0]

        soup = BeautifulSoup(data.text)
        links = [l.get('href') for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        data = requests.get(f"https://fbref.com{links[0]}")
        shooting = pd.read_html(data.text, match="Shooting")[0]
        shooting.columns = shooting.columns.droplevel()

        #Some teams have no shooting stat
        try:
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
        except ValueError:
            continue

        #Filter out other comp
        team_data = team_data[team_data['Comp'] == "Premier League"]

        #Add Team,Season Columns
        team_data['Season'] = year
        team_data['Team'] = team_name
        all_matches.append(team_data)

        #prevent block from scraping website
        time.sleep(1)
    
```

```python
match_df = pd.concat(all_matches)
match_df
```

```python
match_df.to_csv('matches.csv')
```
